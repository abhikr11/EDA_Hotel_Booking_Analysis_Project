#### Chart - 1

````python
#define function to plot pie chart so that we don't have to write code again and again

def plot_pie(data,labels,chart_title):
  '''
  Function to plot pie chart
  '''
  try:
    plt.figure(figsize = (7,7))              #define figure size
    plt.pie(x = data,
            labels = labels,                 #show category labels
            explode=[0.04]* len(data), 
            autopct='%.2f%%', 
            shadow=True)
    plt.title(chart_title, fontsize = 14)
    plt.legend(loc= 'upper right')          #show legend at upper right
    plt.show()                              #show plot

  except:
   print('Invalid Command')
````

````python
#define a function to plot bar chart so that don't have to write code again and again
def plot_bar(x_data, y_data, plot_title, x_label, y_label):
  '''
  Function to plot bar chart
  '''
  try:
    plt.figure(figsize=(10,6))
    sns.barplot(x= x_data,
                y= y_data,
                order= x_data,
                width= 0.6,
                color = 'orange')
    plt.title(plot_title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()

  except:
   print('Invalid Command')
````

````python
#value count
booking_cancelled = df.is_canceled.value_counts()
booking_cancelled
````

````python
#plot pie chart
plot_pie(booking_cancelled,['Not Cancelled', 'Cancelled'],'Percentage of Bookings Cancelled')
````

````python
#hotel wise percentage of bookings canceled and not canceled
#count total booking canceled and not canceled for each hotel
hotel_booking_percent = df.groupby(['hotel', 'is_canceled']).agg({'lead_time':'count'}).reset_index().rename(columns = {'lead_time':'count'})

#calculate percentage of booking canceled and not canceled from count and round off to two decimal
hotel_booking_percent['Percentage_Booking(%)'] = round((hotel_booking_percent['count']/hotel_booking_percent['count'].sum()) * 100, 2)

#replace 0 with not_canceled and 1 with canceled
hotel_booking_percent['is_canceled'].replace({0:'not_canceled', 1:'canceled'}, inplace=True)

#show booking percent DataFrame
hotel_booking_percent
````

````python
#plot bar chart
plt.figure(figsize = (7,5))
sns.barplot(x= 'hotel',
            y='Percentage_Booking(%)',
            hue= 'is_canceled',
            data = hotel_booking_percent,
            palette = 'gist_ncar_r')
plt.legend()
plt.show()   #show plot
````

````python
#total booking for different year
total_booking_yearwise = df.arrival_date_year.value_counts().rename('total_booking')
#total cancellation for different years
total_cancel_yearwise = df[df.is_canceled == 1].arrival_date_year.value_counts().rename('total_cancel')
#percentage cancellation year wise
percent_cancel_yearwise = pd.concat([total_booking_yearwise, total_cancel_yearwise], axis=1).sort_index()
percent_cancel_yearwise['percent_cancel'] = round((percent_cancel_yearwise['total_cancel']/percent_cancel_yearwise['total_booking']) * 100, 2)
#print dataframe
percent_cancel_yearwise
````

````python
#plot year wise number of cancellation
plot_bar(percent_cancel_yearwise.index,
         percent_cancel_yearwise.percent_cancel,
         'Total cancellation over year',
         'Year',
         'Percent cancellaton(%)')
````

````python
#define finction to check same room or not
def alloted_room_check(reserved, assigned):
  '''
  Function to identify if reserved room type and 
  assigned room type is same or different.
  '''
  if reserved == assigned:
    return 'Same'

  else:
    return 'Different'
````

````python
#apply above function to check the room type same or different
df['is_same'] = df.apply(lambda x: alloted_room_check( x['reserved_room_type'], x['assigned_room_type']), axis=1)
````
